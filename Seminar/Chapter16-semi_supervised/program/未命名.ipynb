```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# データ数
n = 200
a = np.linspace(0, math.pi, n / 2)

u_zenhan = np.array([math.cos(i) + 0.5 if i < n/2 else math.cos(i) - 0.5 for i in range(n)]).reshape(n, 1)
u_kouhan = np.random.normal(loc=0, scale=1, size=200).reshape(n, 1)
u = -10 * u_zenhan + u_kouhan

v_zenhan = np.array([math.sin(i) if i < n/2 else -math.sin(i) for i in range(n)]).reshape(n, 1)
v_kouhan = u_kouhan
v = 10 * v_zenhan + v_kouhan
```

```python
x = np.hstack([u, v])
```

```python
y = np.zeros((200, 1))
y[0], y[n-1] = 1, -1
# xの各行の二乗和を求める
x2 = np.power(x, 2)
x2 = np.sum(x2, axis=1)

# ガウスカーネルの幅
hh = 2 * pow(1, 2)
```

```python
x2.reshape(200, 1)
```

```python
x2 = x2.reshape(200, 1)
```

```python
k = np.exp(-np.tile(x2, (1, n)) + np.tile(np.rot90(x2), (n, 1)) - 2 * np.dot(x, np.rot90(x)) / hh)
w = k
```

```python
t = np.dot(np.linalg.inv(np.power(k, 2) + 1 * np.identity(n) + 
                          10 * np.dot(k, np.dot(np.diag(np.sum(w, axis=0) - w), k))),  
           np.dot(k, y))
```

```python
m = 100
X = np.rot90(np.linspace(-20, 20, m).reshape(1, m))
X2 = np.power(X, 2)
U = np.exp(-(np.tile(np.power(u, 2), (1, m)) + np.tile(np.rot90(X2), (n, 1)) - 2 * np.dot(u, np.rot90(X))) / hh)
V = np.exp(-(np.tile(np.power(v, 2), (1, m)) + np.tile(np.rot90(X2), (n, 1)) - 2 * np.dot(v, np.rot90(X))) / hh)
```

```python
fig = plt.figure(figsize=(9, 4))
ax1 = fig.add_subplot(121)
ax1.set_title("laplas")
contour = ax1.contourf(X, X, np.sign(np.dot(np.rot90(V), np.dot(U, np.tile(t, (1, m))))))

print(type(contour))
```
